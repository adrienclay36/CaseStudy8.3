## Q10 Produce a list of facilities with a total revenue less than 1000:

```python
df = pd.read_sql_query("SELECT name, revenue FROM (SELECT f.name, SUM(CASE WHEN b.memid = 0 THEN f.guestcost * b.slots ELSE f.membercost * b.slots END) AS revenue FROM Bookings AS b INNER JOIN Facilities AS f ON f.facid = b.facid GROUP BY f.name) AS subquery WHERE revenue < 1000 ORDER BY revenue", engine)
```

```python
df
```

## Q11: Produce a list of members and who recommended them ordered by surname, firstname

```python
df2 = pd.read_sql_query("SELECT m.firstname, m.surname, (r.surname || ' ' || r.firstname) AS recommendedby FROM Members as m INNER JOIN Members AS r ON m.memid = r.recommendedby ORDER BY m.surname, m.firstname;", engine)
```

```python
df2
```

## Q12 Find facilities with their usage by Member but not guests:

```python
df3 = pd.read_sql_query("SELECT f.name, b.facid, COUNT(b.facid) as MemberBookFreq FROM Bookings as b INNER JOIN Facilities as f ON b.facid = f.facid WHERE b.memid > 0 GROUP BY b.facid", engine)
```

```python
df3
```

## Q13 Find Facilities usage by month, but not guests

```python
df5 = pd.read_sql_query("SELECT f.name AS FacilityName, f.facid as FACID, strftime('%m', starttime) AS Month, COUNT(b.facid) as MemberBookFreq FROM Bookings as b INNER JOIN Facilities AS f ON b.facid = f.facid WHERE b.memid > 0 GROUP BY f.facid UNION ALL SELECT f.name AS FacilityName, f.facid as FACID, strftime('%m', starttime) AS Month, COUNT(b.facid) as MemberBookFreq FROM Bookings as b INNER JOIN Facilities AS f ON b.facid = f.facid WHERE Month = '08' AND b.memid > 0 GROUP BY f.facid UNION ALL SELECT f.name AS FacilityName, f.facid as FACID, strftime('%m', starttime) AS Month, COUNT(b.facid) as MemberBookFreq FROM Bookings as b INNER JOIN Facilities AS f ON b.facid = f.facid WHERE Month = '09' AND b.memid > 0 GROUP BY f.facid", engine)
```
